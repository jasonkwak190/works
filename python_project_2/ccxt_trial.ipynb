# Bithumb pump alert moving average

```python
btc = pybithumb.get_ohlcv("BTC")
close = btc['close']
#print(btc)
#print(close)

# print((close[0] + close[1] + close[2] + close[3] + close[4]) / 5)
# print((close[1] + close[2] + close[3] + close[4] + close[5]) / 5)
# print((close[2] + close[3] + close[4] + close[5] + close[6]) / 5)
# Instead of writing all of them, we use rolling, mean method
window = close.rolling(5) #create 5 days window, group the data by 5 days
ma5 = window.mean()
# we can make it in to one line -> ma5 = close.rolling(5).mean 
last_ma5 = ma5[-2]
print(ma5)

price = pybithumb.get_current_price("BTC")
#Pump? or Dump?
if price > last_ma5:
    print("Pump")
else:
    print("Dump")
```

```python
#Make it into def
def bull_market(ticker):
    df = pybithumb.get_ohlcv(ticker)
    ma5 = df['close'].rolling(window=5).mean()
    price = pybithumb.get_current_price(ticker)
    last_ma5 = ma5[-2]
    
    if price > last_ma5:
        print("Pump")
    else:
        print("Dump")
```

```python
#Try eth to see the function works
is_bull = bull_market("ETH")
if is_bull:
    print("Pump")
```
